```python
import pandas as pd
import numpy as np
```

```python
a=pd.read_csv(r"D:\csv files\Brain Stroke.csv")
a
```

```python
from sklearn.preprocessing import LabelEncoder
l1=LabelEncoder()
```

```python
a['gender']=l1.fit_transform(a['gender'])
a['ever_married']=l1.fit_transform(a['ever_married'])
a['work_type']=l1.fit_transform(a['work_type'])
a['Residence_type']=l1.fit_transform(a['Residence_type'])
a['smoking_status']=l1.fit_transform(a['smoking_status'])
```

```python
a
```

```python
x=a.iloc[:, :-1].values
y=a.iloc[:, -1].values
```

### spliting the dataset into train and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

### feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(x_train)
X_test=sc.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
