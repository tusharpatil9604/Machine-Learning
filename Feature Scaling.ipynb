```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("D:\csv files\Brain Stroke.csv")
```

```python
dataset
```

```python
dataset["ever_married"]=dataset["ever_married"].map({'Yes':1,'No':0})
dataset["work_type"]=dataset["work_type"].map({'Private':0,'Self-employed':1,'Govt_job':2,'children':3})
dataset["Residence_type"]=dataset["Residence_type"].map({'Urban':0,'Rural':1})
dataset["smoking_status"]=dataset["smoking_status"].map({'formerly smoked':0,'never smoked':1,'smokes':2,'Unknown':3})
```

```python
dataset
```

```python
dataset["gender"]=dataset["gender"].map({'Male':1,'Female':0})
```

```python
dataset
```

```python
x=dataset.iloc[:, :-1].values
y=dataset.iloc[:, -1].values
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scale=scaler.fit_transform(x)
```

```python
print(scale)
```

```python
from sklearn.model_selection import train_test_split
scale_train,scale_test,y_train,y_test=train_test_split(scale,y,test_size=0.2,random_state=9)
```

```python
from sklearn.linear_model import LogisticRegression
re=LogisticRegression()
re.fit(scale_train,y_train)
```

```python
y_pred=re.predict(scale_test)
y_pred
```
